```python
import pandas as pd 
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_female', 'sex_male',
       'smoker_no', 'smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
!pip install xgboost 
from xgboost import XGBRegressor
#regressor=RandomForestRegressor(n_estimators=50,criterion='squared_error',max_features='log2',random_state=0)
regressor=XGBRegressor(objective='reg:squarederror', random_state=123)
#regressor=RandomForestRegressor(n_estimators=10,criterion='poisson',random_state=0)
regressor.fit(X_train, Y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="finalized_model_Adaboost.sav"
pickle.dump(regressor,open(filename,'wb'))
```
